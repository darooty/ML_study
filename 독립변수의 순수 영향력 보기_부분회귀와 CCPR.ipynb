```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X0, y = boston['data'], boston['target']
```

```python
import statsmodels.api as sm
```

```python
X = pd.DataFrame(X0, columns=boston['feature_names'])
X = sm.add_constant(X)
X
```

```python
y = pd.DataFrame(y, columns=['target'])
y
```

```python
model = sm.OLS(y, X)
rlt = model.fit()
```

```python
fig = plt.figure(figsize=(20, 20))
sm.graphics.plot_partregress_grid(rlt, fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 20))
sm.graphics.plot_ccpr_grid(rlt, fig=fig)
plt.show()
```
