```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X0, y = pd.DataFrame(boston['data'], columns=boston['feature_names']), pd.DataFrame(boston['target'], columns=['MEDV'])
```

```python
df = sm.add_constant(pd.concat([X0, y], axis=1))
df.head()
```

```python
list(df.columns[:-1].values)
model = sm.OLS.from_formula('MEDV ~ +const+CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT', data=df)
```

```python
rlt = model.fit()
```

```python
rlt.summary()
```

```python
sm.stats.anova_lm(rlt, typ=2)
```

```python
sm.stats.anova_lm(rlt)
```
